```python
# hanoi commander

line = 1

# N개의 disk, M개의 rods.
try :
    N, M = input().split()
    tmp_n = str(int(N))
    tmp_m = str(int(M))

    # 딱 정수형태의 입력만 허용
    if tmp_n != N or tmp_m != M :
        raise Exception('InputIntegerError : N or M')
     어차피 done에서 종료가 될 것임을 알기 때문에  어차피 done에서 종료가 될 것임을 알기 때문에 
    # 예외 발생하지 않았을때
    N = int(N)
    M = int(M)

    # 양수만 입력해야함
    if N>=1 and M>=1 : pass
    else : raise Exception('InputRangeError : N or M')

    # n개의 disk가 rod 1에 크기 오름차순으로 쌓여져 있음
    rod0 = []
    for n in range(N) :
        rod0.append(n)

    # hanoi 0번째 성분 [0,1,2,...N-1] 위에 있을수록 앞에 있음.
    hanoi = [rod0]
    for m in range(M-1) :
        hanoi.append([])

    line += 1
    #print(hanoi)

except Exception as e :
    # 예외 발생했을 때 실행됨.
    print('Failure in Line %d' %line)
    #print(e)

else :
    # 예외 발생하지 않았을때만 일어남.
    pass

finally :
    # 예외 발생 여부와 상관없이 항상 일어남.
    pass 



command = 'start'
while( command != 'done' ) :

    if line == 1 :
        break

    try : 
        text = input().split()

        # 적절한 command 입력
        command = text[0]

        if command == 'move' :
            
            # text 입력이 올바르게 되었는지 확인
            #print(len(text))
            if len(text) != 3 :
                raise Exception('CommandError : input != 3')
            
            # text 입력이 올바름.
            else :
                rod1 = text[1]
                rod2 = text[2]
                #print('rod1 : %s' %rod1)
                #print('rod2 : %s' %rod2)

                # 딱 정수형태의 입력만 허용
                if str(int(rod1)) != rod1 or str(int(rod2)) != rod2 :
                    raise Exception('InputIntegerError : rod1 or rod2')
                rod1 = int(rod1)
                rod2 = int(rod2)
                
                # rod1이 1~M 사이 값을 가지는지
                if 1<=rod1 and rod1<=M : pass
                else : raise Exception('RangeError : rod1')

                # rod2가 1~M 사이 값을 가지고 rod1과 겹치지 않는지
                if 1<=rod2 and rod2<=M and rod1!=rod2 : pass
                else : raise Exception('RangeError : rod2')

            
        elif command == 'done' :

            # text 입력이 올바르게 되었는지 확인
            if  len(text) != 2 :
                raise Exception('CommandError : input != 2')

            # text 입력이 올바름.
            else :
                rod = text[1]

                # 딱 정수형태의 입력만 허용
                if str(int(rod)) != rod :
                    raise Exception('InputIntegerError : rod')
                rod = int(rod)

                # rod가 1~M 사이 값을 가지는지
                if 1<=rod and rod<=M : pass
                else : raise Exception('RangeError : rod')


        # command가 move도 아니고 done도 아니면 에러.
        else :
            raise Exception('CommandError : input ''move'' or ''done''!')


        # 위의 코드까지, command type을 올바르게 받았다면 아래 코드 실행 가능.
        #print('Successful command type')

        # rod1의 가장 위에 올려진 disk를 rod2로 이동
        # hanoi i번째 성분 : [0,1,2,...N-1] 위에 있을수록 앞에 있음.
        if command == 'move' :
            
            # 동일한 디스크에 옮기는 경우, (이미 제외됨.)
            if rod1 == rod2 :
                raise Exception('RodError : same rod')

            # 옯길 disk가 없는 경우
            if len(hanoi[rod1-1]) == 0 :
                raise Exception('EmptyDiskError')

            # del과 insert를 통해서 이동하는 거 구현하기.
            # 가장 상위에 있는 disk일수록 index가 작음.
            tmp = hanoi[rod1-1][0]
            #print('tmp : %d' %tmp)
            del hanoi[rod1-1][0]
            #print(hanoi)
            hanoi[rod2-1].insert(0,tmp)
            #print(hanoi)

            # disk를 옮겼는데 오름차순이 아닐 경우
            check_hanoi = hanoi[rod2-1]
            if check_hanoi != sorted(check_hanoi) :
                raise Exception('DiskOrderError')

            # 모든 과정이 정상적으로 끝나면 다음 줄로 넘어감
            line += 1
            #print('Successful move')
            #print(hanoi)

            
        # target에 N개의 모든 disk가 크기 오름차순으로 되어있는지 판단
        elif command == 'done' :

            # 크기 오름차순 확인은 해당 rod에 list를 확인하여 판단하기. 
            check_hanoi = hanoi[rod-1]
            if check_hanoi != sorted(check_hanoi) :
                raise Exception('DiskOrderError')

            # done 수행 후 '모든' disk가 오름차순으로 되어있는지.
            for i in range(len(hanoi)) :
                check_hanoi = hanoi[i]
                if check_hanoi != sorted(check_hanoi) :
                    raise Exception('DiskOrderError')
            
            # 무사히 for loop를 빠져나왔다는 것은 '모든' disk가 오름차순으로 되어 있다는 뜻!
            print('Success in Line %d' %line)
            break
        

    except Exception as e:
        # 예외 발생했을 때 실행됨.
        print('Failure in Line %d' %line)
        #print(e)
        break
```
